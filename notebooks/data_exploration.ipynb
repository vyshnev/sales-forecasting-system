```python
#testing out the models

import pandas as pd
import sys, os
# Add the project root directory to the Python path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from src.data_preprocessing import load_and_merge_data, handle_missing_values, handle_categorical_data, handle_lagged_features_na
from src.feature_engineering import create_lagged_features, add_time_features
from src.models import train_sarima, predict_sarima, train_prophet, predict_prophet, evaluate_model, split_data

# Load and preprocess the data
data_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'data'))
print(f"Data path is: {data_path}")
df = load_and_merge_data(data_path)
if df is not None:
    # Handle missing values (using 0 for now)
    df = handle_missing_values(df, method='fillna_0')
    # Handle categorical data
    df = handle_categorical_data(df)

    # Feature Engineering
    df = add_time_features(df, 'Date')
    df = create_lagged_features(df, 'Weekly_Sales', lags=[1, 2, 3, 4])
    #Handle NaN values in lagged features
    df = handle_lagged_features_na(df)
    # Split the data
    train_df, test_df = split_data(df, test_size=0.2)

    # --- SARIMA ---
    # Select columns for sarima (can't pass all the columns)
    sarima_train_df = train_df[['Date', 'Weekly_Sales']].set_index('Date')
    sarima_test_df = test_df[['Date', 'Weekly_Sales']].set_index('Date')
    order = (1, 1, 1) # Example order (p, d, q)
    seasonal_order = (1, 1, 1, 52) # Example seasonal order (P, D, Q, s), weekly seasonality for 52 weeks
    sarima_model = train_sarima(sarima_train_df, order, seasonal_order)
    sarima_predictions = predict_sarima(sarima_model, len(sarima_test_df))
    sarima_mae, sarima_rmse = evaluate_model(sarima_test_df['Weekly_Sales'], sarima_predictions)

    print(f"\nSARIMA MAE: {sarima_mae:.2f}")
    print(f"SARIMA RMSE: {sarima_rmse:.2f}")

    # --- Prophet ---
    # For prophet we need to have the columns named 'ds' (date) and 'y' (target variable)
    prophet_train_df = train_df.rename(columns={'Date':'ds', 'Weekly_Sales': 'y'})
    prophet_test_df = test_df.rename(columns={'Date':'ds', 'Weekly_Sales': 'y'})
    # Add regressors for prophet, removing features that are not numerical
    regressors = ['Temperature', 'Fuel_Price', 'MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4',
                'MarkDown5','CPI', 'Unemployment', 'Size', 'Type_A', 'Type_B', 'Type_C', 'month', 'year', 'dayofweek', 'weekofyear',
                'Weekly_Sales_lag_1', 'Weekly_Sales_lag_2', 'Weekly_Sales_lag_3', 'Weekly_Sales_lag_4']

    prophet_model = train_prophet(prophet_train_df, regressors=regressors)
    prophet_predictions = predict_prophet(prophet_model, prophet_test_df)
    prophet_mae, prophet_rmse = evaluate_model(prophet_test_df['y'], prophet_predictions['yhat'])

    print(f"\nProphet MAE: {prophet_mae:.2f}")
    print(f"Prophet RMSE: {prophet_rmse:.2f}")
```

```python
import pandas as pd
import sys, os
# Add the project root directory to the Python path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))

from src.data_preprocessing import load_and_merge_data, handle_missing_values, handle_categorical_data, handle_lagged_features_na
from src.feature_engineering import create_lagged_features, add_time_features
from src.models import train_sarima, predict_sarima, train_prophet, predict_prophet, evaluate_model, split_data, create_future_dataframe, forecast_sarima, scale_target_variable, inverse_scale_target_variable
import matplotlib.pyplot as plt

# Load and preprocess the data
data_path = os.path.abspath(os.path.join(os.getcwd(), '..', 'data'))
print(f"Data path is: {data_path}")
df = load_and_merge_data(data_path)
if df is not None:
    # Handle missing values (using 0 for now)
    df = handle_missing_values(df, method='fillna_0')
    # Handle categorical data
    df = handle_categorical_data(df)

    # Feature Engineering
    df = add_time_features(df, 'Date')
    df = create_lagged_features(df, 'Weekly_Sales', lags=[1, 2, 3, 4])
    #Handle NaN values in lagged features
    df = handle_lagged_features_na(df)
    # Split the data
    train_df, test_df = split_data(df, test_size=0.2)
    
    # --- SARIMA ---
    # Select columns for sarima (can't pass all the columns)
    sarima_train_df = train_df[['Date', 'Weekly_Sales']].set_index('Date')
    sarima_test_df = test_df[['Date', 'Weekly_Sales']].set_index('Date').copy()
    order = (2, 1, 2) # Example order (p, d, q)
    seasonal_order = (0, 1, 0, 52) # Example seasonal order (P, D, Q, s), weekly seasonality for 52 weeks
    sarima_model = train_sarima(sarima_train_df, order, seasonal_order)
    sarima_predictions = predict_sarima(sarima_model, len(sarima_test_df), sarima_test_df)
    sarima_mae, sarima_rmse = evaluate_model(sarima_test_df['Weekly_Sales'], sarima_predictions)

    print(f"\nSARIMA MAE: {sarima_mae:.2f}")
    print(f"SARIMA RMSE: {sarima_rmse:.2f}")
    

    # --- Prophet ---
    # For prophet we need to have the columns named 'ds' (date) and 'y' (target variable)
    prophet_train_df, min_sales, max_sales = scale_target_variable(train_df.rename(columns={'Date':'ds', 'Weekly_Sales': 'y'}), 'y')
    prophet_test_df = test_df.rename(columns={'Date':'ds', 'Weekly_Sales': 'y'}).copy()
    prophet_test_df = scale_target_variable(prophet_test_df, 'y')[0]
    # Add regressors for prophet, removing features that are not numerical
    regressors = ['Temperature', 'Fuel_Price', 'MarkDown1', 'MarkDown2', 'MarkDown3', 'MarkDown4',
                'MarkDown5','CPI', 'Unemployment', 'Size', 'Type_A', 'Type_B', 'Type_C', 'month', 'year', 'dayofweek', 'weekofyear',
                'Weekly_Sales_lag_1', 'Weekly_Sales_lag_2', 'Weekly_Sales_lag_3', 'Weekly_Sales_lag_4']

    prophet_model = train_prophet(prophet_train_df, regressors=regressors)
    prophet_predictions = predict_prophet(prophet_model, prophet_test_df)
    prophet_test_df = inverse_scale_target_variable(prophet_test_df, 'y', min_sales, max_sales)
    prophet_mae, prophet_rmse = evaluate_model(prophet_test_df['original_y'], prophet_predictions['yhat'])

    print(f"\nProphet MAE: {prophet_mae:.2f}")
    print(f"Prophet RMSE: {prophet_rmse:.2f}")
    

    # --- Forecasting for Future ---
    # Train on all the data to get full model
    prophet_full_train_df, min_sales_full, max_sales_full = scale_target_variable(df.rename(columns={'Date':'ds', 'Weekly_Sales': 'y'}), 'y')
    full_prophet_model = train_prophet(prophet_full_train_df, regressors=regressors)

    # Generate Future dates for 1 quarter.
    future_dates_df = create_future_dataframe(df, periods = 12, freq='W') # 3 months assuming weekly data, 4 weeks per month
    
    future_dates_df = future_dates_df.rename(columns={'Date':'ds'})
     # Create all the regressors for future dates. This assumes that we can compute the other regressors
    # Note that for this example we will be filling all the future regressors as 0, but we would
    # typically use forecast methods to predict them as well.
    for regressor in regressors:
      future_dates_df[regressor] = 0 #set to 0 for this example
    # Use the trained Prophet model to make predictions on future dates.
    future_prophet_predictions = predict_prophet(full_prophet_model, future_dates_df)
    future_prophet_predictions = scale_target_variable(future_prophet_predictions, 'yhat')[0]
    future_prophet_predictions = inverse_scale_target_variable(future_prophet_predictions, 'yhat', min_sales_full, max_sales_full)

    print("\nFuture Sales Forecasts (Prophet):")
    print(future_prophet_predictions[['ds', 'original_yhat']]) # print the date and predicted values.

    # --- Forecasting for Future SARIMA
    sarima_full_train_df = df[['Date', 'Weekly_Sales']].set_index('Date')
    full_sarima_model = train_sarima(sarima_full_train_df, order, seasonal_order)
    future_sarima_dates = pd.date_range(start=sarima_full_train_df.index[-1], periods=12, freq='W')
    future_sarima_predictions = forecast_sarima(full_sarima_model, future_sarima_dates, sarima_full_train_df )

    print("\nFuture Sales Forecasts (SARIMA):")
    print(future_sarima_predictions) #print the future dates
    
     # --- Visualization for Future Forecasts ---
    plt.figure(figsize=(10, 6))
    plt.plot(future_dates_df['ds'], future_prophet_predictions['original_yhat'], label='Prophet Predictions', color = 'red')
    plt.plot(future_sarima_dates, future_sarima_predictions, label = "SARIMA Predictions", color='blue')
    plt.title('Future Sales Forecasts')
    plt.xlabel('Date')
    plt.ylabel('Weekly Sales')
    plt.legend()
    plt.grid()
    plt.show()
```
